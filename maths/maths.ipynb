```python
"""
name: starting_point
input: Min/Max LP in Canonical Form
LP_std = convert Min/Max LP to standard form without artificial variable
if no initial BFS
    call Big-M(LP) or 2-Phase(LP)
else
    basis = [list of basic variables from initial BFS]
    call generalised_simplex(LP_std, basis)
"""
```

```python
"""
name: generalised_simplex(LP, basis)
input:  Min/Max LP in standard form [with or without artificial variables]
        basis is list of basic variables in sequence of row number

display initial tableau

while True
    # first select entering variable
    if LP is maximisation problem
        candidates = [list of non-basic variables with most negative coefficient in z-row]
    else
        candidates = [list of non-basic variables with most positiive coefficient in z-row]
    # to choose lowest subscript, for the same i, x_i < s_i < a_i where x are input variables, s are slack/surplus variables, a are artificial variables
    entering variable = variable with lowest subscript in candidates

    if all constraints have non-positive coefficients
        raise Unbounded Error 
    else
        # select leaving variable
        candidates = [list of basic variables with lowest ratio]
        # to choose lowest subscript, for the same i, x_i < s_i < a_i where x are input variables, s are slack/surplus variables, a are artificial variables
        leaving variable = variable with lowest subscript in candidates
        # update tableau
        call row_operations(LP, entering variable, leaving variable)
        replace leaving variable with entering variable in the basis list
        display tableau

if there exists a non-zero artificial variable
    raise No Feasible Solution Error
else if there exists non-basic variable with coefficient 0 in z-row
    multiple solution found
else
    Unique solution found
"""
```

```python
""" 
name: row_operations(LP, entering variable, leaving variable)
input: LP in standard form, entering variable, leaving variable
a = column number of entering variable
b = row number of leaving variable
rows = number of rows in LP
coeff = LP[row a][col b]
LP[row a] = LP[row a] / coeff
for i = 0 to rows-1
    if i != a
        coeff = LP[row i][col b]
        LP[row i] = LP[row i] - coeff*LP[row a]
"""
```

```python
"""
name: Big-M(LP)
input: LP in Canonical Form

let M represents a big positive number

LP = Convert LP to Standard Form with Artificial Variables

# update z-row
if LP is minimisation
    assign -M to each a_i in the z-row
else
    assign M to each a_i in the z-row

basis = [list of basic variables from initial BFS]

# remove each a_i from z-row
for each a_i
    row_operations(LP, a_i, a_i)

call generalised_simplex(LP, basis)
"""
```

```python
def isPrime(n:int) -> bool:
    if n < 2:
        return False
    return n == lowest_divisor(n)

def lowest_divisor(n: int) -> int:
    d = 2
    while d*d <= n:
        if n % d == 0:
            return d 
        if d == 2:
            d += 1
        else:
            d += 2 
    return n 

def eq1(x: int, y:int) -> int:
    return abs(3*x**2 - 4*y**2)

def eq2(x: int, y:int) -> int:
    return abs(4*x - 3*y)

def imply(p:bool, q:bool) -> bool:
    return (not p) or q 

rows = [["x", "y", "|3x2 - 4y2|", "|4x - 3y|", "Both Prime implies (x=1) or (y=1)"]]
for x in range(1,20):
    for y in range(1,10):
        e1, e2 = eq1(x,y), eq2(x,y)
        p1, p2 = isPrime(e1), isPrime(e2)
        implication = imply(p1 and p2, x==1 or y == 1)
        if not implication:
            rows.append([x, y, e1, e2, implication])
TableForm(rows).as_latex()
```

```python
def simplex_maximise(vars:list[Symbol], A:Matrix) -> Matrix:
    """
    Assume header list contains all symbols except the RHS
    Assume matrix A contains a basic feasible solution and is in the following format:
    first column represents Z 
    the subsequent columns represent input variables
    the final column represents RHS
    """
    n_rows, n_cols = A.shape
    n_constraints = n_rows - 1
    nbv = [vars[0]] + vars[-n_constraints::]
    header = Matrix([Symbol("")] + vars + [Symbol("rhs")]).transpose()
    rows = [A.row(i) for i in range(n_constraints + 1)]
    ratios = [Symbol("")] * n_rows
    #pretty_print(header.col_join(Matrix(nbv).row_join(Matrix.vstack(*rows))))
    optimal = False 
    while not optimal:
        print()
        # pick entering variable
        min_index = 0
        for i in range(n_cols):
            if rows[0][i] < rows[0][min_index]:
                min_index = i
        if rows[0][min_index] >= 0:
            # no negative coefficeints in row 0
            optimal = True
        else:
            # pick leaving variable
            min_ratio = oo
            leaving = -1
            ratios = [Symbol("_")]
            for i in range(1, n_constraints+1):
                if rows[i][min_index] > 0:
                    ratio = rows[i][-1]/rows[i][min_index]
                    ratios.append(ratio.evalf(5))
                    if ratio < min_ratio:
                        min_ratio = ratio
                        leaving = i
                else:
                    ratios.append(Symbol("_"))
            if leaving == -1:
                # no leaving variable found
                raise Exception("Unbounded error")

            print("entering variable:", vars[min_index], ", leaving variable:", nbv[leaving])

            pretty_print(header.col_join(Matrix(nbv).row_join(Matrix.vstack(*rows))).row_join(Matrix([Symbol("ratio")] + ratios)))

            nbv[leaving] = vars[min_index]
            # perform row operations
            rows[leaving] = rows[leaving] / rows[leaving][min_index]
            for i in range(n_constraints + 1):
                if i != leaving:
                    rows[i] = rows[i] - rows[i][min_index]*rows[leaving]

    final = header.col_join(Matrix(nbv).row_join(Matrix.vstack(*rows)))
    print("final answer")
    pretty_print(final)
    return final

     
```

```python
class Graph():

    def __init__(self, vertices):
        self.V = vertices
        self.graph = [[0 for column in range(vertices)]
                      for row in range(vertices)]
    
    def set(self, v1, v2, weight):
        self.graph[v1][v2] = weight 
        self.graph[v2][v1] = weight

    def printSolution(self, dist):
        print("Vertex \t Distance from Source")
        for node in range(self.V):
            print(node, "\t\t", dist[node])

    # A utility function to find the vertex with
    # minimum distance value, from the set of vertices
    # not yet included in shortest path tree
    def minDistance(self, dist, sptSet):

        # Initialize minimum distance for next node
        min = 1e7

        # Search not nearest vertex not in the
        # shortest path tree
        for v in range(self.V):
            if dist[v] < min and sptSet[v] == False:
                min = dist[v]
                min_index = v

        return min_index

    # Function that implements Dijkstra's single source
    # shortest path algorithm for a graph represented
    # using adjacency matrix representation
    def dijkstra(self, src):

        dist = [1e7] * self.V
        dist[src] = 0
        sptSet = [False] * self.V

        for cout in range(self.V):

            # Pick the minimum distance vertex from
            # the set of vertices not yet processed.
            # u is always equal to src in first iteration
            u = self.minDistance(dist, sptSet)

            # Put the minimum distance vertex in the
            # shortest path tree
            sptSet[u] = True

            # Update dist value of the adjacent vertices
            # of the picked vertex only if the current
            # distance is greater than new distance and
            # the vertex in not in the shortest path tree
            for v in range(self.V):
                if (self.graph[u][v] > 0 and 
                   sptSet[v] == False and 
                   dist[v] > dist[u] + self.graph[u][v]):
                    dist[v] = dist[u] + self.graph[u][v]

        self.printSolution(dist)
```
